```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup 
from selenium import webdriver
import requests_html
import csv
```

```python
advanceentrieslist = pd.read_csv('AdvanceEntry.csv')
updateddf = advanceentrieslist[['uscfid', 'FIDE ID']]
```

```python
URL = 'https://www.vegaschessfestival.com/live/results/'
page = requests.get(URL, headers={'User-Agent': 'Mozilla/5.0'})
soup = BeautifulSoup(page.content, 'lxml')
```

```python
tables = soup.find_all(class_ = 'doublescroll')
string = tables[0].text
listofrows = string.split('\n')
```

```python
placement = []
title = []
name = []
federation = []
fide_id = []
fide_rating = []
uscf_rating = []
score = []
round1 = []
round2 = []
round3 = []
round4 = []
round5 = []
round6 = []
round7 = []
round8 = []
round9 = []
```

```python
for player in listofrows:
    placement.append(player[0:3])
    title.append(player[4:7])
    name.append(player[8:35])
    federation.append(player[35:38])
    fide_id.append(player[39:47])
    fide_rating.append(player[48:53])
    uscf_rating.append(player[54:58])
    score.append(player[59:62])
    round1.append(player[63:67])
    round2.append(player[68:72])
    round3.append(player[73:77])
    round4.append(player[78:82])
    round5.append(player[83:87])
    round6.append(player[88:92])
    round7.append(player[93:97])
    round8.append(player[98:102])
    round9.append(player[103:107])
```

```python
df = pd.DataFrame(list(zip(placement, title,name,federation,fide_id,fide_rating,uscf_rating,score,round1,round2,round3 ,round4 ,round5 ,round6 ,round7 ,round8,round9 )),
               columns =['Place', 'Title', 'Name', 'Federation', 'Fide ID', 'FIDE rating', 'USCF rating', 'Score', 'Round1', 'Round2', 'Round3', 'Round4', 'Round5', 'Round6', 'Round7', 'Round8', 'Round9'])
```

```python
df.drop(index=df.index[0], 
        axis=0, 
        inplace=True)
```

```python
uscflist = updateddf['uscfid'].tolist()
fidelist = updateddf['FIDE ID'].tolist()
```

```python
newlistuscf = []
for item in uscflist:
    new = item.replace('\xa0', '')
    newlistuscf.append(new)
newlistfide= []

for items in fidelist:
    newz = items.replace('\xa0', '')
    newlistfide.append(newz)
```

```python
updateddf['real uscf'] = newlistuscf
updateddf['real fide'] = newlistfide
```

```python
updateddf = updateddf.drop(['uscfid', 'FIDE ID'], axis = 1)
```

```python
joineddf = pd.merge(left=df, right=updateddf, how='left', left_on='Fide ID', right_on='real uscf')
```

```python
N = 2
joineddf.drop(index=joineddf.index[-N:], 
        axis=0, 
        inplace=True)
```

```python
listofbirthyears = []
for item in fideids:
    try:
        URL = 'https://ratings.fide.com/profile/' + item
        page = requests.get(URL, headers={'User-Agent': 'Mozilla/5.0'})
        soup = BeautifulSoup(page.content, 'lxml')
        fidecard = soup.findAll('div', class_ = 'profile-top-info__block__row__data')
        year = fidecard[3].text
        listofbirthyears.append(year)
    except:
        listofbirthyears.append('UNKNOWN')
```

```python
for item in therest:
    try:
        URL = 'https://ratings.fide.com/profile/' + item
        page = requests.get(URL, headers={'User-Agent': 'Mozilla/5.0'})
        soup = BeautifulSoup(page.content, 'lxml')
        fidecard = soup.findAll('div', class_ = 'profile-top-info__block__row__data')
        year = fidecard[3].text
        listofbirthyears.append(year)
    except:
        listofbirthyears.append('UNKNOWN')
```
